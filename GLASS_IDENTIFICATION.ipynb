```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

glass = pd.read_csv("/content/glass.data")
names = ['Id','RI','Na','Mg','Al','Si','K','Ca','Ba','Fe','glass_type']
glass.columns = names
```

```python
features = ['RI','Na','Mg','Al','Si','K','Ca','Ba','Fe']
label = ['glass_type']

X = glass[features]

y = glass[label]
```

```python
x2 = X.values

from matplotlib import pyplot as plt
import seaborn as sns
for i in range(1,9):
        sns.distplot(x2[i])
        plt.xlabel(features[i])
        plt.show()
```

```python
import seaborn as sns
x2 = pd.DataFrame(X)

plt.figure(figsize=(8,8))
sns.pairplot(data=x2)
plt.show()
```
